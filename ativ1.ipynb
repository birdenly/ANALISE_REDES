# QUESTAO 4

Grafos que podem ter mais de uma forma de ser mostrado. exemplo abaixo:



```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
D = nx.Graph()
D.add_nodes_from([1, 2, 3, 4])
D.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

nx.draw(G, with_labels=True)
plt.show()
nx.draw(D, with_labels=True)
plt.show()
```

# QUESTAO 5

```python
#desenho do grafo bipartido
import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()
X = [1, 2, 3, 4]
Y = [5, 6, 7, 8, 9]
G.add_nodes_from(X, bipartite=0)
G.add_nodes_from(Y, bipartite=1)

E =[(1,3),(1,9),(2,6),(2,7),(2,8),(3,4),(3,5),(3,6),(3,7),(4,7),(4,9),(5,1),(5,2),(5,9),(6,3),(8,2),(8,3),(9,7)]
G.add_edges_from(E)

X_nodes = {n for n, d in G.nodes(data=True) if d['bipartite'] == 0}
Y_nodes = set(G) - X_nodes

pos = nx.bipartite_layout(G, X_nodes)
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
#matriz adijacente do grafo bipartido
A = nx.adjacency_matrix(G)
print(A.todense())
```

```python
#calculo do grau do lado esquerdo e depois direito
for n in X:
  print(G.degree(n))
print()
for n in Y:
  print(G.degree(n))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
V = ['M', 'N', 'O', 'P', 'Q', 'R', 'S']
E =[('M', 'S'), ('N', 'O'), ('P', 'R'), ('N', 'S'), ('O', 'M'), ('N', 'Q'), ('O', 'M'), ('P', 'P'), ('S', 'M'), ('O', 'N'),  ('S', 'M'), ('N', 'R'), ('P', 'M'), ('M', 'S')]

G = nx.DiGraph()
G.add_nodes_from(V)
G.add_edges_from(E)
nx.draw(G, with_labels=True)
plt.show()
```

```python
print(nx.is_path(G,('M','S'))) #a
```

```python
for line in nx.simple_cycles(G, length_bound=4): #nao existe

    print(line)
```

```python
print('nx.is_connected(G)') #C - Função não definicada para graficos direcionados, O grafico é conectado.
```

```python
print(G.degree('N'), G.degree('R'))#d
```

# QUESTAO 10

```python
import networkx as nx
import matplotlib.pyplot as plt

V = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
edges = [('B', 'C', {'weight': 2}), ('B', 'D', {'weight': 2}), ('G', 'A', {'weight': 2}), ('D', 'G', {'weight': 1}), ('D', 'E', {'weight': 1}),('D', 'F', {'weight': 1}), ('D', 'H', {'weight': 1}), ('A', 'B', {'weight': 1}), ('C', 'D', {'weight': 1})]

D = nx.Graph()
D.add_nodes_from(V)
D.add_edges_from(edges)

pos = {'A': (0, 0), 'B': (1, 1), 'C': (1, 2), 'D': (2, 1), 'E': (3, 1), 'F': (3, -1), 'G': (4, 0), 'H': (5, 0)}

nx.draw(D, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(D, 'weight')
nx.draw_networkx_edge_labels(D, pos, edge_labels=edge_labels)
plt.show()

```
